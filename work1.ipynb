```python
import pandas as pd
import random
import numpy as np

df_format=pd.DataFrame([['']])
df_new=pd.DataFrame([[]])

print('n×nます計算をします。nの値と 加減乗除 を選んでください')
n=int(input('n：'))
mean=input('[＋][－][*][/]で入力してください:')

#基準
for j in range(n):
  row=random.randint(1,9)
  df_new[0]=int(row)
  df_format=pd.concat([df_format,df_new],ignore_index=True)
for i in range(n):
  line=random.randint(1,9)
  df_format[i+1]=int(line)

#問題用紙
df_q=df_format.copy()
df_q.loc[0,0]=mean
for i in range(1,n+1):
  for j in range(1,n+1):
    df_q.loc[j,i]=''
print('')
df_q=df_q.to_string(header=False,index=False)
print(df_q)

#模範解答
if mean == '+':
  df_ans=df_format.copy()
  for i in range(1,n+1):
    for j in range(1,n+1):
      df_ans.loc[i,j]=df_format.loc[0,j]+df_format.loc[i,0]

elif mean == '-':
  print('')
  print('上の値から左の値を引いてください')
  df_ans=df_format.copy()
  for i in range(1,n+1):
    for j in range(1,n+1):
      df_ans.loc[i,j]=df_format.loc[0,j]-df_format.loc[i,0]

elif mean == '*':
  df_ans=df_format.copy()
  for i in range(1,n+1):
    for j in range(1,n+1):
      df_ans.loc[i,j]=df_format.loc[0,j]*df_format.loc[i,0]

elif mean == '/':
  print('上の値を左の値で割ってください')
  print('整数で出力し、余りは考えないものとします')
  df_ans=df_format.copy()
  for i in range(1,n+1):
    for j in range(1,n+1):
      kazu=df_format.loc[0,j]/df_format.loc[i,0]
      df_ans.loc[i,j]=int(kazu)

#解答用紙
print('')
print('一行ずつ入力していきます')
df_a=df_format.copy()
for i in range(1,n+1):
  for j in range(1,n+1):
    if mean == '+':
      kaitou=int(input(f'{df_format.loc[0,j]}+{df_format.loc[i,0]}='))
      df_a.loc[i,j]=kaitou
    elif mean == '-':
      kaitou=int(input(f'{df_format.loc[0,j]}-{df_format.loc[i,0]}='))
      df_a.loc[i,j]=kaitou
    elif mean == '*':
      kaitou=int(input(f'{df_format.loc[0,j]}*{df_format.loc[i,0]}='))
      df_a.loc[i,j]=kaitou
    elif mean == '/':
      kaitou=int(input(f'{df_format.loc[0,j]}/{df_format.loc[i,0]}='))
      df_a.loc[i,j]=kaitou

#答え合わせ
count=0
for i in range(1,n+1):
  for j in range(1,n+1):
    if df_a.loc[i,j] == df_ans.loc[i,j]:
      count+=1
print('')
print(f'得点は{count}点です')
if count == n*n:
  print('おめでとう。全問正解です')
else:
  print(f'残念、{n*n-count}箇所間違っています')


#df_format
print('')
print('模範解答')
display(df_ans)
#df_q
print('')
print('あなたの解答')
display(df_a)
```
